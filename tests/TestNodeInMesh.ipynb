```python
import pygimli as pg

world = pg.meshtools.createWorld(start=[0, -10], end=[10, 0])
print(world)
scheme = pg.DataContainerERT()
scheme.createSensor([3.2, -7.3])
world.createNode(scheme.sensor(0))
print(world)
mesh = pg.meshtools.createMesh(world, area=1)
pg.show(mesh)
el = scheme.sensor(0)
n = mesh.findNearestNode(el)
print(el, mesh.node(n).pos(), el.distance(mesh.node(n).pos()))
```
