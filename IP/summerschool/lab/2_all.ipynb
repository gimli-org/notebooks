```python
from glob import glob
import matplotlib.pyplot as plt
from pygimli.physics import SIPSpectrum
```

In the folder, there are SIP spectra of four probes, of which three are consolidated (sandstone?):
BR, KE, NU
and one (TM) unconsolidated. Each probe has been measured twice.

We are plotting each pair together into an axis that has been prepared before.

```python
fig, ax = plt.subplots(nrows=2, figsize=(7, 7))
for filename in glob("BR*.res"):
    spec = SIPSpectrum(filename, k=0.0275)
    spec.showData(ax=ax, label=filename[:4]+filename[-5:-4])
fig.savefig("BR-data.pdf", bbox_inches="tight")
```

```python
fig, ax = plt.subplots(nrows=2, figsize=(7, 7))
for filename in glob("KE*.res"):
    spec = SIPSpectrum(filename, k=0.0275)
    spec.showData(ax=ax, label=filename[:4]+filename[-5:-4])
fig.savefig("KE-data.pdf", bbox_inches="tight")
```
