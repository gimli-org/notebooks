```python
from glob import glob
import matplotlib.pyplot as plt
from pygimli.physics import SIPSpectrum
```

```python
fig, ax = plt.subplots(nrows=2, figsize=(7, 7))
for filename in glob("TM*.res"):
    k = 0.0197 if filename.startswith("BR") else 0.0275
    spec = SIPSpectrum(filename, k=k)
    spec.showData(ax=ax, label=filename[:4])
```

```python
spec = SIPSpectrum('KE-02_06-09-23_50mSm_#1.res', k=0.0275)
spec.showData()
```

```python
spec.removeEpsilonEffect()
spec.showAll()
```

```python
spec.showDataKK()
```

```python
spec.cutF(1000)
spec.showData();
```
