## DC/IP inversion

This example uses the ERTIPManager for inverting a field data set presented by Martin et al. (2020).
It has been measured with the ABEM Terrameter LS 2 using 41 electrodes with a spacing of 1m over a alumn shale with strong IP effects. Raw data and processing scripts are openly available on Zenodo (Martin et al., 2021). Apparent resistivity data are in Fig. 4 (top right), the apparent chargeability in Fig. 6 (right column) and the inversion result is in Fig. 10 (3rd column) in the Martin et al. (2020) paper.

```python
from pygimli.physics import ert
```

```python
data = ert.load("schleizTDIP.dat")
ert.show(data)
print(data)
```

```python
_ = data.show("ip")
```

```python
data.estimateError()
mgr = ert.ERTIPManager(data)
mgr.invert(verbose=True)
```

```python
mgr.showResults(reskw={'cMin': 5, 'cMax': 500}, ipkw={'cMax': 200})
```
