```python
import pygimli as pg
```

```python
grid = pg.createGrid(4,4,4)
grid["m"] = pg.utils.grange(1, grid.cellCount(), 1)
```

```python
pl, _ = pg.show(grid, style="wireframe", hold=True)
pg.viewer.pv.drawMesh(pl, grid, label="m", filter={"threshold": dict(value=14, scalars="m", invert=True)})
pg.viewer.pv.drawMesh(pl, grid, label="m", filter={"slice": dict(origin=[1, 1, 1], normal=[1, 1, 0])})
pl.show()
```
