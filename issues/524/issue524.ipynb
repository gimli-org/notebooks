```python
import numpy as np
import pygimli as pg
import pygimli.meshtools as mt
from pygimli.viewer.pv import drawSensors, drawSlice
from pygimli.physics import TravelTimeManager
```

```python
x = np.arange(-300.,301.,50)
y = np.arange(-300.,301.,50)
z = np.arange(-300.,1.,50)
mesh=mt.createMesh3D(x=x, y=y, z=z)
```

```python
vel3 = np.zeros(mesh.cellCount())
pg_z = pg.z(mesh.cellCenters())

for i,j in enumerate(pg_z):
    if j >= -200. and j <= -100. :
        vel3[i] = 2000.
    else:
        vel3[i] = 1000.

pg.show(mesh, vel3, showMesh=True,filter={'clip':{}},alpha=0.8)
```

```python
numberGeophones = 7
paodian= 7
lst2=np.linspace(-300.,301., numberGeophones)
lst1=np.linspace(-300.,301.,paodian)
sensors=[]
for i in lst1:
    sensors.append([i,-250.,0.])
for j in lst2:
    sensors.append([j, 250., 0.])
scheme = pg.physics.traveltime.createRAData(sensors)
mgr = TravelTimeManager()
plotter, _ = pg.show(mesh, style='wireframe', line_width=0.1,
                          hold=True)
drawSensors(plotter,sensors[0:7], diam=20, color='red')
drawSensors(plotter,sensors[7:], diam=20, color='green')
plotter.show()
```

```python
data = mgr.simulate(slowness=1.0 / vel3, scheme=scheme, mesh=mesh,
                     noiseLevel=0.001, noiseAbs=0.001, seed=1337,
                     verbose=True)
vest = mgr.invert(data,mesh=mesh,verbose=True,maxIter=50)
```

```python
ax, _ = pg.show(mesh, alpha=0., hold=True, colorBar=False)
drawSlice(ax,mesh,normal=[1,0,0], data=pg.abs(vest), label="Absolute velocity")
ax.show()
```

```python
ax1, _ = pg.show(mesh, alpha=0., hold=True, colorBar=False)
drawSlice(ax1,mesh,normal=[0,1,0], data=pg.abs(vest), label="Absolute velocity")
ax1.show()
```
