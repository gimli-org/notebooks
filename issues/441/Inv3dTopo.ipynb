```python
import pygimli as pg
import pygimli.meshtools as mt
from pygimli.physics import ert
```

```python
data = ert.load('AllLines.dat')
data.remove(data['rhoa'] < 0.01)
pg.info('Filtered rhoa (min/max)', min(data['rhoa']), max(data['rhoa']))
data["k"] = ert.createGeometricFactors(data, numerical=False)
data['err'] = ert.estimateError(data, absoluteError=0, relativeError=0.03)
```

```python
surf = mt.createParaMeshSurface(data)
pg.show(surf)
```

```python
plc = mt.createParaMeshPLC3D(data)
mesh = mt.createMesh(plc)
pg.show(mesh)
```

```python
mgr = ert.ERTManager(data)
mod = mgr.invert(mesh=mesh, verbose=True)
```
