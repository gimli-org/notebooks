# Issue 588 - Neumann BC outcome

This notebook was made on response of a github issue for pyGIMLi.
It discusses the use of the inhomogeneous Neumann boundary condition, which
is obviously put on the flow (a grad u) and not the field (grad u).
An appropriate change of code (to make it work on flow or maybe both by a
switch) or documentation is necessary.
The full issue can be found on https://github.com/gimli-org/gimli/issues/588

```python
import pygimli as pg
from numpy import linspace
```

```python
grid = pg.createGrid(x=linspace(0., 1., 11), y=linspace(0., 1., 11))
pg.show(grid, markers=True)
```

```python
u = pg.solve(grid, a=[0, 50.], bc={'Dirichlet': {1: 0.0}, 'Neumann': {2: 1.0}})
# u = pg.solve(grid, a=10., bc={'Dirichlet': {1: 0.0}, 'Neumann': {2: 1.0}})
pg.viewer.show(grid, u, nLevs=10);
```

```python
gradU = pg.solver.grad(grid, u)
pg.show(grid, gradU[:, 0])
```
